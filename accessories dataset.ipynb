```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the distribution of the 'Sell' prices
plt.figure(figsize=(10, 6))
sns.histplot(data['Sell'], kde=True)
plt.title('Distribution of Sell Prices')
plt.xlabel('Sell Price')
plt.ylabel('Frequency')
plt.show()

# Bar plot of items by color
plt.figure(figsize=(12, 8))
sns.countplot(y='Color 1', data=data, order=data['Color 1'].value_counts().index)
plt.title('Count of Items by Primary Color')
plt.xlabel('Count')
plt.ylabel('Primary Color')
plt.show()
```

```python
# Scatter plot of 'Buy' vs 'Sell' prices
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Buy', y='Sell', data=data)
plt.title('Buy vs Sell Prices')
plt.xlabel('Buy Price')
plt.ylabel('Sell Price')
plt.show()

# Box plot of 'Sell' prices by 'Catalog' status
plt.figure(figsize=(12, 8))
sns.boxplot(x='Catalog', y='Sell', data=data)
plt.title('Sell Prices by Catalog Status')
plt.xlabel('Catalog Status')
plt.ylabel('Sell Price')
plt.show()
```

```python

# Load the dataset
data = pd.read_csv(file_path)

# Count the number of accessories for each primary color
color_counts = data['Color 1'].value_counts().reset_index()
color_counts.columns = ['Primary Color', 'Count']

# Create a horizontal bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y='Primary Color', data=color_counts, palette='viridis')

# Add titles and labels
plt.title('Accessories Sorted by Primary Color')
plt.xlabel('Count')
plt.ylabel('Primary Color')

# Show the plot
plt.show()
```

```python
print(data.dtypes)
```

```python
data['Buy'] = pd.to_numeric(data['Buy'], errors='coerce')
```

```python
data = data.dropna(subset=['Buy'])
```

```python


# Calculate the average buy price for each primary color
average_buy_prices = data.groupby('Color 1')['Buy'].mean().reset_index()

# Sort the colors by the average buy price
average_buy_prices = average_buy_prices.sort_values(by='Buy', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Color 1', y='Buy', data=average_buy_prices, palette='viridis')

# Add titles and labels
plt.title('Average Buy Prices by Primary Color (Sorted)')
plt.xlabel('Primary Color')
plt.ylabel('Average Buy Price')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python

# Calculate the average sell price for each primary color
average_sell_prices = data.groupby('Color 1')['Sell'].mean().reset_index()

# Sort the colors by the average sell price
average_sell_prices = average_sell_prices.sort_values(by='Sell', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Color 1', y='Sell', data=average_sell_prices, palette='viridis')

# Add titles and labels
plt.title('Average Sell Prices by Primary Color (Sorted)')
plt.xlabel('Primary Color')
plt.ylabel('Average Sell Price')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python


# Count the number of accessories for each source
source_counts = data['Source'].value_counts().reset_index()
source_counts.columns = ['Source', 'Count']

# Create a horizontal bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y='Source', data=source_counts, palette='viridis')

# Add titles and labels
plt.title('Count of Accessories by Source')
plt.xlabel('Count')
plt.ylabel('Source')

# Show the plot
plt.tight_layout()
plt.show()
```

```python


# Count the number of accessories for each style
style_counts = data['Style'].value_counts().reset_index()
style_counts.columns = ['Style', 'Count']

# Create a horizontal bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y='Style', data=style_counts, palette='coolwarm')

# Add titles and labels
plt.title('Count of Accessories by Style')
plt.xlabel('Count')
plt.ylabel('Style')

# Show the plot
plt.tight_layout()
plt.show()
```

```python


# Calculate the average buy price for each style
average_buy_prices = data.groupby('Style')['Buy'].mean().reset_index()

# Sort the styles by the average buy price
average_buy_prices = average_buy_prices.sort_values(by='Buy', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Style', y='Buy', data=average_buy_prices, palette='viridis')

# Add titles and labels
plt.title('Average Buy Prices by Style')
plt.xlabel('Style')
plt.ylabel('Average Buy Price')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python

# Calculate the average sell price for each style
average_sell_prices = data.groupby('Style')['Sell'].mean().reset_index()

# Sort the styles by the average sell price
average_sell_prices = average_sell_prices.sort_values(by='Sell', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Style', y='Sell', data=average_sell_prices, palette='viridis')

# Add titles and labels
plt.title('Average Sell Prices by Style')
plt.xlabel('Style')
plt.ylabel('Average Sell Price')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python


# Calculate the average buy price for each source
average_buy_prices = data.groupby('Source')['Buy'].mean().reset_index()

# Sort the sources by the average buy price
average_buy_prices = average_buy_prices.sort_values(by='Buy', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Source', y='Buy', data=average_buy_prices, palette='pastel')

# Add titles and labels
plt.title('Average Buy Prices by Source')
plt.xlabel('Source')
plt.ylabel('Average Buy Price')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python


# Calculate the average sell price by source
avg_sell_price_by_source = data.groupby('Source')['Sell'].mean().reset_index()

# Sort the values for better visualization
avg_sell_price_by_source = avg_sell_price_by_source.sort_values(by='Sell', ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Sell', y='Source', data=avg_sell_price_by_source, palette='pastel')

# Add title and labels
plt.title('Average Sell Prices by Source')
plt.xlabel('Average Sell Price')
plt.ylabel('Source')

# Show the plot
plt.show()
```
